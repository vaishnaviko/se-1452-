```python
import matplotlib.pyplot as plt

def quadratic_temperature_model(day, a, b, c):
    # Quadratic equation for temperature
    temperature = a * day**2 + b * day + c
    return temperature

def main():
    # Fixed coefficients for the quadratic equation: ax^2 + bx + c
    a = 5
    b = 3
    c = 4

    # Fixed number of days to model
    num_days = int(input("Enter the number of days to model: "))

    # Lists to store day and corresponding temperature values
    days = list(range(1, num_days + 1))
    temperatures = [quadratic_temperature_model(day, a, b, c) for day in days]

    # Plotting
    plt.plot(days, temperatures, label=f'Temperature: {a}x^2 + {b}x + {c}')
    plt.title('Weather Modeling')
    plt.xlabel('Day')
    plt.ylabel('Temperature (Â°C)')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

def quadratic_temperature_model(day, a, b, c):
    # Quadratic equation for temperature
    temperature = a * day**2 + b * day + c
    return temperature

def main():
    # Fixed coefficients
    fixed_a, fixed_b, fixed_c = 2, 3, 5

    # Get user input for quadratic coefficients
    user_a = float(input("Enter the quadratic coefficient (a): "))
    user_b = float(input("Enter the linear coefficient (b): "))
    user_c = float(input("Enter the constant term (c): "))

    # Get user input for the number of days to model
    num_days = int(input("Enter the number of days to model: "))

    # Lists to store day and corresponding temperature values
    days = list(range(1, num_days + 1))

    # Calculate temperatures for both fixed and user-input coefficients
    fixed_temperatures = [quadratic_temperature_model(day, fixed_a, fixed_b, fixed_c) for day in days]
    user_temperatures = [quadratic_temperature_model(day, user_a, user_b, user_c) for day in days]

    # Plotting
    plt.plot(days, fixed_temperatures, label=f'Fixed Coefficients: {fixed_a}x^2 + {fixed_b}x + {fixed_c}')
    plt.plot(days, user_temperatures, label=f'User Coefficients: {user_a}x^2 + {user_b}x + {user_c}')
    plt.title('Weather Modeling')
    plt.xlabel('Day')
    plt.ylabel('Temperature (Â°C)')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
def quadratic_weather_model(time, a, b, c):
      temperature = a * (time ** 2) + b * time + c
      return temperature

def main():
      print("Quadratic Weather Modeling")
      print("==========================")

      try:
          with open('multi.txt', 'r') as file:
              lines = file.readlines()

          time_values = list(range(0, 11))
          plt.figure(figsize=(8, 6))

          for line in lines:
              coefficients = line.split()
              a, b, c = map(float, coefficients)

              temperature_values = [quadratic_weather_model(t, a, b, c) for t in time_values]

              plt.plot(time_values, temperature_values, marker='o', linestyle='-', label=f'a={a}, b={b}, c={c}')

          plt.title('Temperature Variation Over Time')
          plt.xlabel('Time')
          plt.ylabel('Temperature')
          plt.grid(True)
          plt.xlim(0, 10)
          plt.legend()
          plt.show()

      except FileNotFoundError:
          print("File not found. Please make sure 'multi.txt' exists.")

if __name__ == "__main__":
      main()
```
